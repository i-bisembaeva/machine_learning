```python
import re
import numpy as np
data = []
with open ('sentences.txt', 'r') as file:
    for line in file:
        lowstr = line.lower()
        data.append(lowstr)
for i in range(len(data)):
    data[i] = re.split('[^a-z]', data[i])
    for j in range(data[i].count('')):
        data[i].remove('')
#print(data)
dict = {}
count = 0
for line in data:
    for word in line:
        if not (word in dict.keys()):
            dict[word] = count
            count += 1
#print(dict)
matrix = np.zeros((len(data), len(dict)))
for i in range(len(data)):
    for j in range(len(dict)):
        for item in dict.items():
            if item[1] == j:
                break
            else:
                matrix[i][j] = data[i].count(item[0])
print(matrix)
np.set_printoptions(threshold=np.inf)
print(matrix)
```
